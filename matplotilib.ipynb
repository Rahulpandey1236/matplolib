```python
x = [1, 2, 3, 4, 5]
y = [5, 1, 7, 8, 2]

plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x,y,color = "red")
plt.title("plot line to understand the chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
plt.plot(x,y,color = "red", marker = "o")
plt.title("thisa is syntax")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
          
```

```python
plt.plot(x, y, color = "red", marker = "o", linestyle = '--')
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x, y, color = "red", marker = "o", linestyle = '--', linewidth = 3)
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x, y, color = "red", marker = "o", linestyle = '--', linewidth = 3, markersize=10)
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x, y, color = "red", marker = "o", linestyle = '--', linewidth = 3, markersize=10)
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [5, 1, 7, 8, 2]
y2 = [7, 8, 3, 5, 1]

plt.plot(x,y1,color = "blue",marker = "o",markersize = 12,linestyle='--',linewidth=3,label = "line y1")
plt.plot(x,y2,color = "red",marker = "o",markersize = 12,linestyle='--',linewidth=3,label = "line y2")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()

plt.title("this is syntx ")
plt.show()
```

```python
x = np.linspace(0,10,100)
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.plot(x, np.sin(x), '--')
```

```python
plt.plot(x,np.cos(x),'--')
```

```python
plt.plot(x, np.sin(x), '.-')
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.plot(x,np.sin(x))


plt.subplot(1,2,2)
plt.plot(x,np.cos(x))

plt.show()
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))


plt.subplot(2,2,3)
plt.plot(x,np.tan(x))

plt.subplot(2,2,4)
plt.plot(x,np.cos(x+1))

plt.show()
```

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))


plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))


plt.show()
```

```python
plt.plot(x,np.sin(x-0),color = "pink")
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.axis('equal')
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid()
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)

plt.title("basic syntax")
plt.xlabel("x label")
plt.ylabel("y label")
plt.axis('equal')
plt.grid()
plt.show()
```
