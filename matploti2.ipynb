```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = ['a', 'b', 'c', 'd']
y = [4, 6, 1, 2]


plt.bar(x, y, color = 'r')
```

```python
x = ['a','b','c','d']
y = [1,2,3,4]

plt.barh(x,y,color= "r")
```

```python
#pie chart >> To understand the percnt dist of data
langs = ['C', 'Python', 'Java']
students = [20, 100, 40]


plt.pie(students, labels = langs, autopct = '%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
explode = (0.0, 0.1, 0.1)

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]


plt.pie(students, labels = langs, autopct = '%1.1f%%', explode = explode, shadow = True)
plt.title("Pie chart")
plt.show()
```

```python
#3d plot

x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)


fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x, y, z, c='red')
```
